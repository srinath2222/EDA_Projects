```python
# Plotting countplots for distribution of data in each column.
```

```python
plt.figure(figsize =(10,6) )

sns.set_style('whitegrid')
sns.countplot(x=Data['Age'], native_scale = True)
plt.title('Age Distribution')
```

```python
sns.set_style('whitegrid')
sns.countplot(Data, x=Data['VisitFrequency'],hue='Gender', native_scale = True)
plt.title('Visiting Frequency Distribution')
```

```python
sns.set_style('whitegrid')
sns.countplot(Data , x=Data['PreferredCuisine'],hue = 'Gender', native_scale = True)
plt.title('Preffered Cuisine Distribution')
```

```python
sns.set_style('whitegrid')
sns.countplot(Data,x=Data['TimeOfVisit'],hue='Gender',native_scale = True)
plt.title('Time Of Visit Distribution')
```

```python
sns.set_style('whitegrid')
sns.countplot(Data,x=Data['GroupSize'],hue='Gender')
plt.title('Group Size Distribution')
```
