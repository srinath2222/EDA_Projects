### Visualization

```python
# Lets us plot a countplot to show the count of age groups.

sns.countplot(x='Age',data=Data,native_scale=True)
```

```python
# To plot a countplot to show the count of 'male' and 'female' in a specific age groups in our dataset.

sns.set_style(style='whitegrid')
sns.countplot(x='Age',hue = 'Gender',data=Data,native_scale=True) # in gender '0' is male and '1' is female
```

```python
# To show how many DiscountsAvailed based on on gender.

sns.set_style('whitegrid')
sns.countplot(x='DiscountsAvailed', hue='Gender' ,data=Data)
```

```python
# To plot a countplot for Annualincome

sns.set_style('whitegrid')
sns.displot(Data, y="AnnualIncome", bins=30)
```

```python
Data.head()
```

```python
fig = plt.figure(figsize =(10,6) )
sns.barplot(data= Data, x= 'Gender',y= 'NumberOfPurchases',hue = 'ProductCategory', palette='dark')
plt.ylabel('Number Of Purchases')
plt.xlabel('Gender')
plt.title('Distribution of Product Category by Gender');
plt.show()
```

```python
# plotting a pie chart on product category

Data['ProductCategory'].value_counts()
```

```python
data = [289, 331, 273, 321, 321]
keys = ['0 - Electronics', '1 - Clothing', '2 - Home Goods', '3 - Beauty', '4 - Sports']

plt.pie(data, labels=keys, autopct='%.0f%%', startangle = 90,counterclock=False,shadow=True)
plt.title('Product Category')
plt.show() 
```

```python
# plotting a pie chart for product purchace status ( will customer purchase or not. its like a prediction )

Data['PurchaseStatus'].value_counts()
```

```python
values = [852,648]
labels = ['0 - No Purchase', '1 - Purchase']

plt.pie(values, labels=labels, autopct='%.0f%%', startangle = 90,counterclock=False)
plt.title('Purchase Status')
plt.show() 
```
