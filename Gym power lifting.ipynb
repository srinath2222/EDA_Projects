```python
Data = Data.drop_duplicates(inplace= False)
```

```python
# categorizing the data and their count.

Data['Lifter Name'].value_counts()
```

```python
Data['Weight Class'].value_counts()
```

```python
Data['Lift Type'].value_counts()
```

```python
Data['Amount Lifted (kg)'].value_counts()
```

```python
# lifters whose age is more than 50.

Data.loc[Data['Age'] > 50].sort_values('Age') # sorted ages more than 50
```

```python
# pie chart on lift types.

values = [1025,1011,964]
lift_type = ['Deadlift', 'Bench Press','Squat']

plt.pie(values, labels=lift_type, autopct='%.0f%%', startangle = 90,counterclock=False)
plt.title('Lift Type')
plt.show() 
```

```python
# To check is any outliers are present in amount of weight lifted based on age.

plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
sns.boxplot(data=Data, x='Age', y='Amount Lifted (kg)', color='pink')

# There are no outliers present in the data
```

```python

plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
sns.boxplot(data=Data, x='Weight Class', y='Amount Lifted (kg)',hue='Lift Type', native_scale = True)
plt.title('Box Plot ')
```

```python

plt.figure(figsize=(6,4))
sns.set_style('whitegrid')
sns.boxplot(data=Data, x='Lift Type', y='Amount Lifted (kg)')
plt.title('Box Plot ')
```

```python
fig = plt.figure(figsize =(8,5) )
sns.countplot(Data,x='Lifter Name' ,native_scale=True , color = 'skyblue')
plt.xticks(rotation=90)
```

```python
fig = plt.figure(figsize =(8,5) )
sns.countplot(Data,x='Lifter Name' , hue='Lift Type', native_scale=True)
plt.xticks(rotation=90)
```
