```python
# Visualization - 
```

```python
columns = ['IMDb Rating','Year','Certificates','Genre','MetaScore']

for col in columns:
    plt.figure(figsize=(14,6))
    sns.countplot(movie_list_1,x=col)
    plt.xticks(rotation=90)
    plt.show()
```

```python
sns.lineplot(movie_list_1,x='Genre',y='IMDb Rating')
sns.set_style('darkgrid')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(movie_list_1,x='Genre',y='IMDb Rating')
sns.set_style('darkgrid')
plt.xticks(rotation=90)
plt.show()
```

```python
cols = ['IMDb Rating','Year','MetaScore','Duration (minutes)']

sns.heatmap(movie_list_1[cols].corr(),annot=True,fmt='.2g')
plt.show()
```

```python
sns.pairplot(movie_list_1)
sns.set_style('darkgrid')
plt.show()
```
